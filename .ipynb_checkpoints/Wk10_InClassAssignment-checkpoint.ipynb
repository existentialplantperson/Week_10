#### 1. Write a function that calculates the dot product manually AND a function that calculates the cross product manually. Both should accept two vectors and perform that calculation. Be sure to use try except and error handling for basic issues that might occur.

```python
#define function to calculate dot product
def dot(a,b):
    
    #variable will be updated
    results = 0
    
    try:
    #create for loop
        for i in range(len(a)) and range(len(b)):
            results = results + (a[i] * b[i])
            
    #create exception for index error
    except IndexError:
        return 'Arrays must be of the same shape'
    
    #create exception for type error
    except TypeError:
        return 'Arrays must contain only numbers'
    
    #create exception for any remaining errors
    except Exception as e:
        return e
    
    #if no exceptions return the results
    else:
        return results
        
```

```python
a = [3.6,4.8, 5.2]
b = [4,5,6]

dot(a,b)
```

#### 2. In a markdown section in your notebook, explain what one hot encoding is and why we use it. Then, manually calculate a one hot encoding for the Belize_Airbnb_listings.csv neighborhood_group column. Note: manually means donâ€™t use scikitlearn or keras.

one hot encoding converts categorical data into numerical data which is important for machine learning and certain types of data manipulation.

```python
import pandas as pd
import numpy as np
from numpy import argmax

belize_df = pd.read_csv("belize_airbnb_listings.csv")
```

```python
belize_df.head(1)
```

```python
#get unique values from column 
#iterate through list and create new column for each column
#one way, new column for each neighborhood
#convert list to series and append to list
#another way, one column with an array
```

```python
#find unique neighborhoods in data set
groups = belize_df['neighbourhood_group'].unique()
print(groups)
group_series = belize_df['neighbourhood_group']

#map each neighborhood group to an integer
group_to_int = dict((g, i) for i, g in enumerate(groups))
int_to_group = dict((i, g) for i, g in enumerate(groups))

# integer encode input data
integer_encoded = [group_to_int[data] for data in group_series]

# one hot encode, for each value in the integer_encoded list
onehot_encoded = list()
for value in integer_encoded:
    
    #create a matrix for each neighborhood group (6)
    neighborhood = [0 for _ in range(len(groups))]
    neighborhood[value] = 1
    onehot_encoded.append(neighborhood)

# invert encoding
inverted = int_to_group[argmax(onehot_encoded[0])]
print(inverted)

#convert encoded list into a series
encoded_series = pd.Series(onehot_encoded)

#merge encoded series onto original df as new column
belize_df2 = pd.concat([belize_df, encoded_series], axis="columns")
belize_df2.head(5)
```
